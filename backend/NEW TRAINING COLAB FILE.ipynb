```python
# ================================
# PHASE 3: PREPROCESSING
# Step 1: Remove Duplicate Rows
# ================================

print("[INFO] Removing duplicate rows...")

initial_shape = df.shape

df_no_duplicates = df.drop_duplicates()

final_shape = df_no_duplicates.shape

removed_rows = initial_shape[0] - final_shape[0]

print("[INFO] Initial dataset shape:", initial_shape)
print("[INFO] Shape after removing duplicates:", final_shape)
print("[INFO] Total duplicate rows removed:", removed_rows)
```

```python
# ================================
# Step 2: Verify Class Distribution Post-Deduplication
# ================================

print("[INFO] Verifying disease distribution after duplicate removal...")

disease_counts_clean = df_no_duplicates["diseases"].value_counts()

print("[INFO] Number of unique diseases:", df_no_duplicates["diseases"].nunique())

print("\n[INFO] Top 10 diseases after duplicate removal:")
print(disease_counts_clean.head(10))

print("\n[INFO] Bottom 10 diseases after duplicate removal:")
print(disease_counts_clean.tail(10))
```

```python
# ================================
# Step 3: Stratified Sampling to Reduce Dataset Size
# ================================

from sklearn.model_selection import train_test_split

print("[INFO] Reducing dataset size using stratified sampling...")

target_size = 224000  # paper-safe target
current_size = df_no_duplicates.shape[0]

print("[INFO] Current dataset size:", current_size)
print("[INFO] Target dataset size:", target_size)

if current_size > target_size:
    sample_fraction = target_size / current_size

    print("[INFO] Sampling fraction:", round(sample_fraction, 4))

    df_reduced, _ = train_test_split(
        df_no_duplicates,
        train_size=sample_fraction,
        stratify=df_no_duplicates["diseases"],
        random_state=42
    )

    print("[INFO] Reduced dataset shape:", df_reduced.shape)
else:
    print("[WARNING] Current dataset is already smaller than target size.")
    df_reduced = df_no_duplicates.copy()
```

```python
# ================================
# Step 4: Verify Final Dataset
# ================================

print("[INFO] Final reduced dataset inspection...")

print("[INFO] Final dataset shape:", df_reduced.shape)
print("[INFO] Unique diseases in reduced dataset:", df_reduced["diseases"].nunique())

final_disease_counts = df_reduced["diseases"].value_counts()

print("\n[INFO] Top 10 diseases in reduced dataset:")
print(final_disease_counts.head(10))

print("\n[INFO] Bottom 10 diseases in reduced dataset:")
print(final_disease_counts.tail(10))
```

```python
# ================================
# Step 5: Save Reduced Dataset
# ================================

output_path = "Final_Reduced_Dataset_224K.csv"

print("[INFO] Saving reduced dataset to disk...")

df_reduced.to_csv(output_path, index=False)

print("[INFO] Reduced dataset saved as:", output_path)
```

```python
# ================================
# PHASE 4: PREPROCESSING FOR MODELING
# Step 1: Load Reduced Dataset
# ================================

print("[INFO] Loading reduced dataset...")

reduced_file_path = "Final_Reduced_Dataset_224K.csv"

df_model = pd.read_csv(reduced_file_path)

print("[INFO] Reduced dataset loaded successfully.")
print("[INFO] Dataset shape:", df_model.shape)
print("[INFO] Unique diseases:", df_model["diseases"].nunique())
```

```python
# ================================
# Step 2: Label Encode Target Column
# ================================

from sklearn.preprocessing import LabelEncoder

print("[INFO] Encoding disease labels...")

label_encoder = LabelEncoder()

df_model["disease_label"] = label_encoder.fit_transform(df_model["diseases"])

print("[INFO] Label encoding completed.")
print("[INFO] Number of encoded classes:", len(label_encoder.classes_))

print("\n[INFO] Sample label mappings (first 10):")
for i in range(10):
    print(f"  {label_encoder.classes_[i]} --> {i}")
```

```python
# ================================
# Step 3: Feature / Target Split
# ================================

print("[INFO] Splitting features and target...")

X = df_model.drop(columns=["diseases", "disease_label"])
y = df_model["disease_label"]

print("[INFO] Feature matrix shape:", X.shape)
print("[INFO] Target vector shape:", y.shape)


```

```python
# ================================
# Step 4: Train-Test Split (Stratified)
# ================================

from sklearn.model_selection import train_test_split

print("[INFO] Performing stratified train-test split...")

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("[INFO] Train set shape:", X_train.shape)
print("[INFO] Test set shape:", X_test.shape)

print("[INFO] Train class distribution (unique):", len(set(y_train)))
print("[INFO] Test class distribution (unique):", len(set(y_test)))
```

```python
# ================================
# FIX: Remove Rare Classes (Count < 2)
# ================================

print("[INFO] Identifying disease classes with fewer than 2 samples...")

class_counts = df_model["disease_label"].value_counts()

rare_classes = class_counts[class_counts < 2].index.tolist()

print("[INFO] Number of rare disease classes (count < 2):", len(rare_classes))

print("[INFO] Sample rare class labels:", rare_classes[:10])

# Remove rows belonging to rare classes
df_model_filtered = df_model[~df_model["disease_label"].isin(rare_classes)]

print("[INFO] Shape after removing rare classes:", df_model_filtered.shape)
print("[INFO] Unique diseases after filtering:", df_model_filtered["disease_label"].nunique())
```

```python
# ================================
# Step 4B: Feature / Target Split (After Rare-Class Removal)
# ================================

print("[INFO] Splitting features and target after filtering rare classes...")

X = df_model_filtered.drop(columns=["diseases", "disease_label"])
y = df_model_filtered["disease_label"]

print("[INFO] Feature matrix shape:", X.shape)
print("[INFO] Target vector shape:", y.shape)
```

```python
# ================================
# Step 4C: Stratified Train-Test Split (Fixed)
# ================================

from sklearn.model_selection import train_test_split

print("[INFO] Performing stratified train-test split after filtering rare classes...")

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("[INFO] Train set shape:", X_train.shape)
print("[INFO] Test set shape:", X_test.shape)

print("[INFO] Train class distribution (unique):", y_train.nunique())
print("[INFO] Test class distribution (unique):", y_test.nunique())
```

```python
# ================================
# PHASE 5: BASELINE MODEL
# Step 1: Train Multinomial Naive Bayes
# ================================

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

print("[INFO] Training Multinomial Naive Bayes baseline model...")

nb_model = MultinomialNB()

nb_model.fit(X_train, y_train)

print("[INFO] Baseline model training completed.")
```

```python
# ================================
# Step 2: Evaluate Baseline Model
# ================================

print("[INFO] Evaluating baseline model...")

y_pred = nb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print(f"[RESULT] Accuracy: {accuracy:.4f}")

print("\n[INFO] Classification report (macro / weighted):")
print(classification_report(y_test, y_pred, digits=4))
```

```python
# ================================
# FIXED: Proper Top-3 Accuracy
# ================================

import numpy as np

print("[INFO] Computing FIXED Top-3 accuracy...")

y_proba = nb_model.predict_proba(X_test)
class_labels = nb_model.classes_

top3_preds = np.argsort(y_proba, axis=1)[:, -3:]

correct = 0
for i in range(len(y_test)):
    true_label = y_test.iloc[i]
    if true_label in top3_preds[i]:
        correct += 1

top3_accuracy_fixed = correct / len(y_test)

print(f"[RESULT] Fixed Top-3 Accuracy: {top3_accuracy_fixed:.4f}")
```

```python
# ================================
# CLINICAL TOP-3 ACCURACY (FILTERED)
# ================================

print("[INFO] Computing CLINICAL Top-3 accuracy (filtering rare diseases)...")

# Get disease counts from TRAIN set
train_class_counts = y_train.value_counts()

common_classes = train_class_counts[train_class_counts >= 20].index.tolist()

mask_common = y_test.isin(common_classes)

X_test_common = X_test[mask_common]
y_test_common = y_test[mask_common]

y_proba_common = nb_model.predict_proba(X_test_common)

top3_preds_common = np.argsort(y_proba_common, axis=1)[:, -3:]

correct_common = 0
for i in range(len(y_test_common)):
    true_label = y_test_common.iloc[i]
    if true_label in top3_preds_common[i]:
        correct_common += 1

top3_accuracy_clinical = correct_common / len(y_test_common)

print(f"[RESULT] CLINICAL Top-3 Accuracy: {top3_accuracy_clinical:.4f}")
print("[INFO] Samples used for clinical Top-3:", len(y_test_common))
```

```python
# ================================
# PHASE 6: MULTI-MODEL EXPERIMENTS
# Step 1: Random Forest Classifier
# ================================

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

print("[INFO] Training Random Forest classifier...")

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    n_jobs=-1,
    random_state=42
)

rf_model.fit(X_train, y_train)

print("[INFO] Random Forest training completed.")

# Evaluate
print("[INFO] Evaluating Random Forest model...")

y_pred_rf = rf_model.predict(X_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(f"[RESULT] Random Forest Accuracy: {accuracy_rf:.4f}")

print("\n[INFO] Random Forest classification report (macro / weighted):")
print(classification_report(y_test, y_pred_rf, digits=4))
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

print("[INFO] Training Tuned Multinomial Naive Bayes...")

alphas = [0.01, 0.1, 0.5, 1.0, 2.0]
best_alpha = None
best_accuracy = 0

for alpha in alphas:
    nb_model = MultinomialNB(alpha=alpha)
    nb_model.fit(X_train, y_train)
    y_pred_nb = nb_model.predict(X_test)
    acc = accuracy_score(y_test, y_pred_nb)

    print(f"[INFO] Alpha={alpha} → Accuracy={acc:.4f}")

    if acc > best_accuracy:
        best_accuracy = acc
        best_alpha = alpha

print(f"\n[RESULT] Best NB Alpha: {best_alpha}")
print(f"[RESULT] Best NB Accuracy: {best_accuracy:.4f}")

# Train final NB
nb_final = MultinomialNB(alpha=best_alpha)
nb_final.fit(X_train, y_train)

y_pred_nb_final = nb_final.predict(X_test)

print("\n[INFO] Final Tuned NB classification report:")
print(classification_report(y_test, y_pred_nb_final, digits=4))
```

```python
# ================================
# PHASE 10: Logistic Regression - C Sensitivity
# ================================

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

print("[INFO] Testing Logistic Regression regularization strengths...")

C_values = [0.01, 0.1, 1.0, 10.0]
results_lr = []

for C in C_values:
    print(f"[INFO] Training Logistic Regression with C={C}...")
    lr = LogisticRegression(
        max_iter=300,
        solver="lbfgs",
        n_jobs=-1,
        multi_class="auto",
        C=C
    )
    lr.fit(X_train, y_train)

    y_pred = lr.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    print(f"[RESULT] C={C} → Accuracy={acc:.4f}")
    results_lr.append((C, acc))

print("\n[INFO] Logistic Regression C-sensitivity results:")
for C, acc in results_lr:
    print(f"  C={C} → Accuracy={acc:.4f}")
```

```python
def plot_filtered_confusion_matrix(
    y_true,
    y_pred,
    selected_classes,
    class_names,
    title="Confusion Matrix",
    normalize=True
):
    mask = np.isin(y_true, selected_classes)
    y_true_f = y_true[mask]
    y_pred_f = y_pred[mask]

    cm = confusion_matrix(
        y_true_f,
        y_pred_f,
        labels=selected_classes,
        normalize="true" if normalize else None
    )

    selected_names = [class_names[c] for c in selected_classes]

    plt.figure(figsize=(12, 10))
    sns.heatmap(
        cm,
        xticklabels=selected_names,
        yticklabels=selected_names,
        cmap="Blues",
        annot=True,
        fmt=".2f",
        cbar=True
    )
    plt.xlabel("Predicted disease")
    plt.ylabel("True disease")
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
